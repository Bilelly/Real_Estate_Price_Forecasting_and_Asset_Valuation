```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset unified
df = pd.read_csv('../data/df_cleaned.csv',sep=';')
```

```python
# Select numerical variables
var_num = df.select_dtypes(include=['int', 'float'])
# Select categorical variables
var_cat = df.select_dtypes(include=['object','category'])
```

```python
# Display descriptive statistics for numerical variables
print("Descriptive Statistics for Numerical Variables:")
display(var_num.describe().T)
# Display descriptive statistics for categorical variables
print("Descriptive Statistics for Categorical Variables:")
display(var_cat.describe().T)
```

```python
plt.figure(figsize=(18, 14))

# 2. Calculate Spearman correlation
corr = var_num.corr(method='spearman')

# 3. Create the mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# 4. Plotting
sns.heatmap(corr, 
            mask=mask, 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm', 
            square=True, 
            linewidths=.5,
            annot_kws={"size": 10}, 
            cbar_kws={"shrink": 0.8})

plt.title('Correlation Matrix for Numerical Variables (Spearman)', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(18, 14))

# 1. On calcule la corrÃ©lation sur var_num en excluant la cible
# On s'assure que 'prix_m2_vente' est retirÃ© s'il Ã©tait prÃ©sent
corr = var_num.drop(columns=['prix_m2_vente'], errors='ignore').corr(method='spearman')

# 2. Masque pour le triangle supÃ©rieur (Ã©vite les doublons)
mask = np.triu(np.ones_like(corr, dtype=bool))

# 3. Affichage du Heatmap
sns.heatmap(corr, 
            mask=mask, 
            annot=True, 
            fmt=".2f", 
            cmap='RdBu_r', # 'RdBu_r' est excellent pour voir les corrÃ©lations opposÃ©es
            center=0,      # Force le blanc Ã  0
            square=True, 
            linewidths=.5,
            cbar_kws={"shrink": 0.8})

plt.title('Inter-corrÃ©lation des variables numÃ©riques (Hors Cible)', fontsize=16, fontweight='bold')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# 1. On travaille sur les features uniquement (on exclut la cible)
features_df = var_num.drop(columns=['prix_m2_vente'], errors='ignore')

r2_results = []
cols = features_df.columns

for col in cols:
    # X = toutes les variables sauf 'col', y = 'col'
    X = features_df.drop(columns=[col])
    y = features_df[col]
    
    model = LinearRegression()
    model.fit(X, y)
    r2 = model.score(X, y)
    
    # Calcul du VIF : 1 / (1 - RÂ²)
    # Un VIF > 5 ou 10 indique une forte multicollinÃ©aritÃ©
    vif = 1 / (1 - r2) if r2 < 1 else float('inf')
    
    r2_results.append({
        "Variable": col,
        "RÂ²": r2,
        "VIF": vif
    })

r2_data = pd.DataFrame(r2_results).sort_values(by="VIF", ascending=False)
r2_data
```

```python
# Display distribution of numerical variables

n_cols = len(var_num.columns)
n_rows = (n_cols + 3) // 4  # 4 colonnes par ligne
fig_height = max(12, n_rows * 3)  # Hauteur adaptÃ©e au nombre de lignes

plt.figure(figsize=(20, fig_height))
for i, column in enumerate(var_num.columns, 1):
    plt.subplot(n_rows, 4, i)
    sns.histplot(var_num[column], kde=True, bins=30, color='#339af0', stat='density')
    plt.title(f'Distribution de {column}', fontsize=11, fontweight='bold')
    plt.xlabel(column, fontsize=10)
    plt.ylabel('DensitÃ©', fontsize=10)

plt.tight_layout(pad=3)
plt.show()
```

```python
from scipy.stats import kstest
# KS Test for Normality
results = []

for col in var_num.columns:
    data = df[col].dropna()

    # Normalization 
    data_norm = (data - data.mean()) / data.std()

    # KS test vs normal distribution
    ks_stat, p_value = kstest(data_norm, 'norm')

    # Conclusion on normality
    is_normal = "Yes" if p_value >= 0.05 else "No"

    results.append([col, ks_stat, p_value, is_normal])

# Final results DataFrame
ks_df = pd.DataFrame(results, columns=["Variable", "KS_statistic", "p_value", "Is_Normal"])
ks_df
```
